1. Which Country has Highes and Lowest Happiness Score

```python

happy_score = df.groupby("Country or region")["Score"].sum()
print(happy_score.nlargest(1))
happy_score.nsmallest(1)
```

2. Is there any Correlation between GDP per Capita and Happiness Score:

```python
Correlation = df["GDP per capita"].corr(df["Score"])
Correlation
```

Visual Representation of correlation between GDP and Happiness Score:

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
sns.scatterplot(data=df,x="GDP per capita",y="Score",color="pink")
plt.title("Correlation between GDP and Happiness Score")
plt.xlabel("GDP")
plt.ylabel("Hapiness Score")
```

3. How does social support impart happiness score across the world:

```python
social_support = df.groupby("Social support")["Score"].mean()
social_support
```

Visualization of Happiness Score with Socail Support by scatter plot

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,6))
sns.scatterplot(x="Social support",y="Score",data=df,hue="Score")
plt.title("Visualization of Happiness Score with Socail Support by scatter plot")
plt.xlabel("Social Support")
plt.ylabel("Happiness Score")
plt.tight_layout()
plt.show()
```

4.Does Health Life Expectancy have a significant influence on happiness:

```python
Correlation_health_life = df["Healthy life expectancy"].corr(df["Score"])
Correlation_health_life
```

Correlation between Health Life Expectancy and Happiness Score is "0.77" which indicate moderate to strong postive relationship.

Visualization of correlation between Helthy Life Expectancy and Happiness Score:

```python
sns.set_style("white")
plt.figure(figsize=(9,6))
data= df["Healthy life expectancy"].corr(df["Score"])
sns.heatmap([[data]],vmin=0,vmax=1,annot=True,cmap="coolwarm",xticklabels=["Score"],yticklabels=["Helthy Life Expectancy"])
plt.title("Correlation between Healthy Life Expectancy and Happiness Score")
```

5. Which factor (GDP per capita, Social Support, Healthy Life Expectancy) contributes most to Happiness:

```python
factor = ["GDP per capita","Social support","Healthy life expectancy"]
multiple_correlation = df[factor+["Score"]].corr()["Score"].drop("Score")
multiple_correlation
```

From the above results we can know that GDP per capita factor contributes more to Happiness Score

Visualtion of Which factor (GDP per capita, Social Support, Healthy Life Expectancy) contributes most to Happiness:

```python
plt.figure(figsize=(9,6))
sns.heatmap(multiple_correlation.to_frame(),annot=True,cbar=True,linewidths=1)
plt.title("Correaltion of Factor with Happiness Score")
plt.show()
```

9. Are wealthier counties (higher GDP) consistenly happier:

```python
plt.figure(figsize=(9,6))
df["GDP per capita"] = pd.cut(df["GDP per capita"],bins=[1.0,1.3,1.6,2.0],labels=["low","medium","high"])
sns.boxplot(x="GDP per capita",y="Score",data=df)
plt.title("GDP per capitavs.Happiness Score")
plt.xlabel("GDP per capita")
plt.xticks(rotation=45)
plt.ylabel("Happiness Score")
plt.show()
```

Results:
While wealthier countries generally have higher scores, some exceptions exist (e.g., Qatar with high GDP but moderate happiness).

10.What regional trends can be observed in data:

```python
regional_trends = df.groupby("Country or region")["Score"].mean()
print(regional_trends.nlargest(5))
print(regional_trends.nsmallest(5))
```

Nordic countries have the highest average happiness scores, while sub-Saharan African countries have lower averages.

Conculsion:
1.GDP Per Capita and Happiness:
A strong positive correlation indicates that wealthier countries generally experience higher happiness levels. However, economic prosperity alone is not sufficient for ensuring happiness, as seen in some exceptions where high GDP does not translate into high happiness scores.
2.Social Support:
Social support emerges as one of the most significant contributors to happiness. Countries with strong social support systems consistently report higher happiness scores, highlighting the importance of community and family ties in well-being.
3.Freedom to Make Life Choices:
Freedom to make personal life decisions positively influences happiness. Societies where individuals feel empowered to make their own choices tend to report higher levels of well-being.
4.Perceptions of Corruption:
Perceptions of corruption negatively affect happiness. Countries perceived as less corrupt report higher happiness levels, emphasizing the importance of trust in institutions and governance for well-being.
5.Regional Trends:
Nordic countries consistently rank among the happiest, reflecting a balance of economic prosperity, social support, and trust in governance. Conversely, regions facing economic and political instability, such as sub-Saharan Africa and parts of the Middle East, report lower happiness levels.
6.Key Drivers of Happiness:
Among the factors analyzed, social support and freedom to make life choices appear to have the greatest impact on happiness, followed by GDP per capita and healthy life expectancy. Addressing perceptions of corruption also holds potential for improving happiness.


Policy Implications:
Governments aiming to improve national happiness should prioritize policies that enhance social support networks, ensure access to healthcare, foster economic growth, and promote trust and transparency in governance. Creating environments where individuals feel empowered to make life choices can further enhance well-being.
