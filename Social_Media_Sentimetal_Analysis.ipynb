Data Analysis:

```python
# 1.What is the overall distriution of Sentiment in the dataset:

df["Sentiment"].value_counts()
```

This gives the count of each sentiment category (e.g., Positive, Joy, Excitement, Negative,Neutral...)

```python
# 2. How many unique users in the dataset:

df["User"].nunique()
```

```python
# 3 . Which Platform has the most post:

df["Platform"].value_counts()
```

Shows the number of posts from each platform

```python
# 4. What is the average no. of Likes and Retweet per post:

df[["Likes","Retweets"]].mean()
```

Provides the Average mean value for likes and retweets.

```python
# 5. Which Country has highest number of post:

df["Country"].value_counts().idxmax()
```

Return the highest country with number of posts - USA

```python
# 6. What is the distribution of Post per year:

df["Year"].value_counts().sort_index()
```

Gives the count of post per year

```python
# 7. How does Sentiment vary by Platform:

df.groupby("Platform")["Sentiment"].value_counts().unstack()
```

Creates a pivot table showing the sentiment distribution for each platform.

Counter is imported from Python’s collections module to count occurrences of items efficiently.

```python
# 8. What are the most common Hashtags:

from collections import Counter
hashtags = df["Hashtags"].dropna().str.split()
hashtag_list = [tag for sublist in hashtags for tag in sublist]
Counter(hashtag_list).most_common(10)
```

Finds the top 10 most frequently used hashtags.

```python
# 9. What time of a Day people post most:

df["Hour"].value_counts().sort_index()
```

Shows the number of posts per hour.

```python
# 10. Which Post recevied highest Likes and Retweets:

df.nlargest(5,"Likes")[["Text","Likes"]]
```

```python
df.nlargest(5,"Retweets")[["Text","Retweets"]]
```

Shows the top 5 most liked and retweeted posts by selecting column Text and Likes/Retweets

```python
# 11. What is the average number of Likes and Retweets per sentiment Category:

df.groupby("Sentiment")[["Likes","Retweets"]].mean()
```

* This calculates the average likes and retweets for each sentiment category (Positive, Negative, Neutral).
* Helps in understanding which sentiment tends to get the most engagement.

```python
# 12. Which hour of the day has the highest user engagement(Likes + Retweets):

df.groupby("Hour")[["Likes", "Retweets"]].sum().sum(axis=1).idxmax()
```

axis=1 means summing across columns, i.e., Likes + Retweets for each hour.

* This finds the hour of the day when engagement (likes + retweets) is highest.
* Helps in identifying peak posting hours for maximum visibility.


Data Visualization:

```python
# 11. Distribution of Top 10 Sentiment in the database:

plt.figure(figsize=(8,5))
top_sentiments = df["Sentiment"].value_counts().nlargest(10).index
df_top = df[df["Sentiment"].isin(top_sentiments)]
sns.countplot(x="Sentiment",data = df_top,palette="coolwarm",hue="Platform")
plt.title("Distribution of Sentiments")
plt.xticks(rotation=90)
plt.show()
```

```python
# 12. How do number of Post vary from different Platform:

plt.figure(figsize=(8,4))
sns.countplot(y="Platform",data=df)
plt.title("No.of post vary from different Platforn")
plt.show()
```

```python
# 13. What is the trend of the Post over Time:

df["Timestamp"] = pd.to_datetime(df["Timestamp"])
df["Date"] = df["Timestamp"].dt.date
tweets_per_day = df.groupby("Date").size()
plt.figure(figsize=(10,6))
tweets_per_day.plot(kind="line")
plt.title("Tweet Activity Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.grid(True)
plt.tight_layout
plt.show()
```

```python
# 14. What are the most popular Hashtags:

import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd
hashtag = df["Hashtags"].dropna().str.split(",").explode()
top_hashtags = pd.Series(Counter(hashtag)).nlargest(15)
plt.figure(figsize=(10,8))
top_hashtags.plot(kind="bar",color="skyblue")
plt.title("Top Hashtags")
plt.xlabel("Hashtags")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
 # 15. How does tweet volume vary by country?

Top_Country = df['Country'].value_counts().nlargest(10)
plt.figure(figsize=(8,6))
Top_Country.plot(kind='bar', color='coral')
plt.title("Top 10 Countries by Tweet Count")
plt.xlabel("Country")
plt.ylabel("Number of Tweets")
plt.grid(False)
plt.show()
```

```python
# 16. At what time of the day are tweets most frequently posted?

sns.histplot(df['Hour'], bins=24, kde=True, color='purple')
plt.title("Tweet Frequency by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Number of Tweets")
plt.show()
```

```python
# 17.How do the number of tweets vary across days of the week?

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Weekday'] = df['Timestamp'].dt.day_name()
plt.figure(figsize=(8,6))
sns.countplot(x='Weekday', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='coolwarm',hue="Weekday")
plt.title("Tweets by Day of the Week")
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
# 18.Is there a correlation between likes and retweets?

plt.figure(figsize=(8,8))
sns.scatterplot(x='Likes', y='Retweets', data=df, alpha=0.5)
plt.title("Correlation Between Likes and Retweets")
plt.xlabel("Likes")
plt.ylabel("Retweets")
plt.grid(False)
plt.show()
```
